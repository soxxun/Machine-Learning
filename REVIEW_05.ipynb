### **퍼셉트론 : 인공신경망 중 가장 발달된 모델**

#### 퍼셉트론(신경망) --> 딥러닝(깊은 신경망)
#### 다층 퍼셉트론(얕은 신경망) - 신경망 + 신경망

<br>

퍼셉트론은 입력층과 출력층(한개의 노드)으로 구성된다.
이때 입력층은 'd(특정 벡터의 차원)+1'개의 노드로 이루어져 있다.

iris를 예로 들면 sepal length(X1), sepal width(X2), petal length(X3), petal width(X4) 각각이 특징 벡터가 된다.

<br>

캡처.JPG


퍼셉트론의 연산은 Xi와 Wi를 곱해서 출력 노드에 전달한다.

i번째 입력 노드와 출력 노드는 가중치 Wi를 가진 에지로 표현하고
활성화 함수(커널 트릭)를 이용하여 비선형 분류기로 만들어준다.

X0는 bias가 1인 노드이다. 그리고 이러한 노드를 모두 더하여 s를 계산하고 활성화 함수를 적용한다.

이때 활성화 함수 --> **계단 함수** (s > 0-->1  but   --> -1 )

퍼셉트론의 특징인 특정 벡터를 1또는 -1로 변환하는 이진 분류기(binary classfiyer)

<br>


**<계단함수의 종류>**

- sigmoid : 0 ~ 1 (이진 분류)
- tanh
- relu : sigmoid와 tanh가 갖는 'gradient vanishing' 문제를 해결하기 위한 활성화 함수로, x가 0보다 크면 기울기가 1인 직선, 0보다 작으면 함수 값이 0이 된다. 이는 0보다 작은 값들에서 뉴런이 죽을 수 있는 단점을 야기한다. 또한 sigmoid, tanh 함수보다 학습이 빠르고, 연산 비용이 적고, 구현이 매우 간단하다는 특징이 있다.

- softmax : -1 ~ 1 (다중 분류)

<br>

활성화 함수 적용 이유 : 뉴런을 모방하기 위해, 즉 활성화 함수는 뉴런에서 다음 뉴런으로 전달해주는 역할을 한다. 이때 모든 정보를 전달하는 것이 아니라 적용 함수(sigmoid / softmax / relu 등)에 따라 차이가 있다.




### **활성화 함수**

딥러닝 네트워크에서 노드에 입력된 값들을 비선형 함수에 통과시킨 후 다음 레이어로 전달하는데, 이 때 사용하는 함수를 활성화 함수(Activation Function)라고 한다.

선형 함수가 아니라 비선형 함수를 사용하는 이유는 딥러닝 모델의 레이어 층을 깊게 가져갈 수 있기 때문이다.

인공신경망에서 활성화 함수는 입력 데이터를 다음 레이어로 어떻게 출력하느냐를 결정하는 역할이기 때문에 매우 중요하다.

즉, 활성화 함수는 입력을 받아서 활성화 또는 비활성화를 결정하는 데에 사용되는 함수이다.

#### 가중치 w0 = -0.5, w1=1 w2=1 을 식에 대입하면
#### o = w1*x1 +  w2*x2 + w0 = x1 + x2 - 0.5
#### 0 = x1 + x2 -0.5 
#### x2 = 0.5 - x1
#### 퍼셉트론은 특징 공간을 두 부분의 공간으로 분할하는 이진 분류기

```python
import numpy as np

# 퍼셉트론을 파이썬 형태로
x2 = np.array([1,0,0])

# 가중치 벡터
w = np.array([-0.5, 1, 1])

# 결과 (요소별로 합을 구함)
s = sum(x2 * w)
s

# ex. 활성화 함수를 지정해 주어 0보다 크면 1, 0보다 작으면 -1
if s > 0:
  print(1)
else:
  print(-1)
```
