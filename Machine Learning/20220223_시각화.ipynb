```python
x = np.linspace(-1,1,100) # -1과 1 사이 백개의 값 생성
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)
# 화면 분할
fig, ax = plt.subplots(2,2) # 값은 항상 두 개(fig, ax)
ax[0,0].plot(x,y_1)
ax[0,1].plot(x,y_2)
ax[1,0].plot(x,y_3)
ax[1,1].plot(x,y_4)
plt.show()
```

```python
x_1 = range(100)
y_1 = range(100)

x_2 = range(100)
y_2 = [value + 100 for value in x_2] # 100만큼 차이남

plt.plot(x_1,y_1,c='r', linestyle = 'dashed') # c : color 둘 다 혼용
plt.plot(x_2,y_2,c='b', linestyle = 'dotted')
plt.title("Tow Lines")
plt.show()
```

```python
# 범례 legend
# - shadow 매개변수로 그림자 효과
# - loc 매개변수로 위치
    # - center upper right...... (총 11개)
    # - best 최적의 위치에 자동으로 설정
```

```python
# 이전 그래프에 범례 삽입하기
x_1 = range(100)
y_1 = range(100)

x_2 = range(100)
y_2 = [value + 100 for value in x_2] # 100만큼 차이남

plt.plot(x_1,y_1,c='r', linestyle = 'dashed', label='line_1') # c : color 둘 다 혼용
plt.plot(x_2,y_2,c='b', linestyle = 'dotted', label='line_2') # 범례명도 생성


# 범례
plt.legend(
    shadow=True,
    fancybox = False,
    loc = "best"
)

plt.title("Tow Lines")
plt.xlabel('x_line')
plt.ylabel('y_line')

plt.show()
```

```python
# 막대 그래프 : 데이터의 개수나 크기 비교 (이산형 데이터, 범주형 데이터)
# ex - 서울에 사는 인구 수

data = [
    [5,25,50,20],[4,23,51,17],[6,22,52,19]
]
x = np.arange(0,8,2) # x좌표 설정 0 2 4 6

# 세 개의 막대그래프 생성
plt.bar(x, data[0], color ='r', width = 0.45) # width를 주어서 겹치지 않게 만들기
plt.bar(x + 0.5, data[1], color ='g', width = 0.45)
plt.bar(x + 1.0, data[2], color ='b', width = 0.45)

# x축에 표시될 이름 설정
plt.xticks(x + 0.5, ("a","b","c","d")) # 공간의 개수가 동일해야 함
plt.show()
```

```python
# boxplt : 이상치를 찾아낼 때 사용...4분위수
# - 4분위수 (데이터를 4등분으로 나눴을 때 기준이 되는 수)
#   - 데이터를 균등하게 나눌 수 있는 수
#   - ex) 12를 4로 나누었을 때 3씩 나누어짐

data= np.random.randn(100,5) # 정수로 표현
data
plt.boxplot(data) # 박스에는 균등하게 나누어진 수 50퍼센트의 구간
                  # o <-- 라인과 박스에 속하지 못하는 범위 외의 수
                  # 여러 개의 상황을 동시에 볼 수 있다는 것이 장점
                  #   - ex. 지역별 주거 만족도
                  #         박스가 높을수록 만족도 higher
plt.show()
```
