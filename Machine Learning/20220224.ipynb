```python
# one hot encoding
# 0과 1로 표현
# 성별이 m f로 표현 --> m:0 f:1
# ex. 만족도(높음 증간 낮음) --> (세 가지를 이진수로 표현) 100, 010, 001
# ex. 나이구간(유아 어린이 청소년 청년 중년 장년 노년)
```

```python
edges = pd.DataFrame({
    'source':[0,1,2],
    'target':[2,2,3],
    'weight':[3,4,5],
    'color' :['red', 'blue', 'green'],
})
edges
```

```python
edges.dtypes
```

```python
pd.get_dummies(edges) # one hot encoding을 자동으로 해주는 함수

# red, green, blue....와 같은 문자열은 특정 카테고리가 있는 범주이기 때문에 숫자로 표현하는 것이 아니라 연속형 데이터로 표현
```

```python
# weight 데이터가 숫자로 되어 있어서 연속형처럼 보이지만... 범주형....
# M L XL
```

```python
# edges.loc[edges.loc[:,'weight'] == 3]['weight'] --> 이렇게 하면 컬럼 자체를 변경 불가
# edges.loc[0,'weight'] = 10 --> 이렇게 바꿔줄 수 있게끔 만들기
# edges[edges['weight']== 3] # 데이터프레임
```

```python
# 키값 넣어보기
weight_dic = {
    3:'M',
    4:'L',
    5:'XL'
}
weight_dic
```

```python
print(edges['weight'].map(weight_dic))
print(edges['weight'])
edges['weight'] = edges['weight'].map(weight_dic)
edges
```

```python
pd.get_dummies(edges['weight'])
```

```python
temp = pd.get_dummies(edges['color'])
temp
```

```python
edges = pd.concat([edges,temp], axis=1)
edges
```

```python
# binding(바인딩)
# - 연속형 데이터 --> 범주형 데이터 (변경)
```

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks',
'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons',
'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts',
'Scouts'],
 'company': ['1st', '1st', '2nd', '2nd', '1st',
'1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
 'name': ['Miller', 'Jacobson', 'Ali', 'Milner',
'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger',
'Riani', 'Ali'],
 'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3,
2, 3],
 'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57,
62, 70, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['regiment',
'company', 'name', 'preTestScore', 'postTestScore'])
df
```

```python
bins = [0,25,50,75,100] # 0~25  25~50  50~60    cut 함수
group_name = ['low','okay','good','great'] # postTestScore을 cut함수로 자르기

categories = pd.cut(df['postTestScore'], bins, labels=group_name)
categories
```

```python
# 특정 값을 특정 값으로 변경할 때는 dic으로 만들어서 pd의 map을 적용하면 범주형 데이터로 처리 가능
# 구간이 있는 범주들을 일괄적으로 처리할 때(bins)의 경우에는 pd의 cut함수 사용
```

```python
# caling 데이터의 크기를 맞춰
# 10 30 120 360 1080 - 3배씩
# 100 200 300 400 500 - 2배씩
# - 변동폭 동일하다는 것을 프로그램상으로 인식시키기
# - 변동폭이 다를 때는 변동폭이 큰 쪽을 기준으로 삼음

# 몸무게(40~120)와 키(140~200)

# 범위를 통일 0~100   40~200
# 0 ~ 1 (최대값 최소값을 기준으로 정규화하는 방법)
# (공식) : (xi - min(x)) / (max(x) - min(x))............
# 이러한 공식을 통한 정규화를 z-score라고 함
```
